# Demo 01 — Video-level analysis without global corpus

This notebook shows the simplest end-to-end flow for one video:

1. scrape comments into Bronze
2. preprocess into Silver
3. compute basic stats
4. compute TF-IDF using only the video's own comments

```python
VIDEO_ID = "dQw4w9WgXcQ"
LIMIT = 10000
TOP_X = 50 # top n tokens / keywords to return
NGRAM_MIN = 1
NGRAM_MAX = 2
```

Scrape comments into Bronze.

```python
%%capture
!py -m yt_comments scrape {VIDEO_ID} --limit {LIMIT}
```

Preprocess into Silver.

```python
%%capture
!py -m yt_comments preprocess {VIDEO_ID}
```

Compute basic stats.

```python
!py -m yt_comments stats {VIDEO_ID} --top-n {TOP_X}
```

Compute TF-IDF using only the video's own comments.

```python
!py -m yt_comments tfidf {VIDEO_ID} --top-k {TOP_X} --ngram-min {NGRAM_MIN} --ngram-max {NGRAM_MAX}
```

Artifacts are now available under Bronze, Silver, and Gold in the `data/` folder.
