# Demo 03 — channel-level and distinctive keywords

This notebook shows the channel workflow:

1. discover channel videos
2. scrape channel comments
3. preprocess channel videos
4. compute channel-level stats and TF-IDF
5. compute distinctive keywords for one selected video
6. print final report of all discovered artifacts

```python
CHANNEL_ID = "@kurzgesagt"
VIDEO_ID = "h3DCdWyb0cc" # will be used for distinctive keywords
PUBLISHED_AFTER = "2026-01-01"
LIMIT = 10000
TOP_X = 20
NGRAM_MIN = 1
NGRAM_MAX = 2
```

Discover channel videos.

```python
!py -m yt_comments discover-videos {CHANNEL_ID} --published-after {PUBLISHED_AFTER}
```

Scrape and preprocess channel videos.

```python
%%capture
!py -m yt_comments scrape-channel {CHANNEL_ID} --published-after {PUBLISHED_AFTER} --comments-limit {LIMIT}
!py -m yt_comments preprocess-channel {CHANNEL_ID}
```

Compute channel-level basic stats.

```python
!py -m yt_comments stats-channel {CHANNEL_ID} --top-n {TOP_X}
```

Compute channel-level TF-IDF.

```python
!py -m yt_comments tfidf-channel {CHANNEL_ID} --top-k {TOP_X} --ngram-min {NGRAM_MIN} --ngram-max {NGRAM_MAX}
```

Compute distinctive keywords for one selected video.

```python
%%capture
!py -m yt_comments tfidf {VIDEO_ID} --top-k {TOP_X} --ngram-min {NGRAM_MIN} --ngram-max {NGRAM_MAX}
```

```python
!py -m yt_comments distinctive-keywords {CHANNEL_ID} {VIDEO_ID}
```

Print channel-level final report.

```python
!py -u -m yt_comments report-channel {CHANNEL_ID} --video {VIDEO_ID}
```

Artifacts are now available under Bronze, Silver, and Gold in the `data/` folder.
