# Demo 02 — Video-level analysis with global corpus

This notebook builds a global corpus artifact and then re-runs TF-IDF for one video using corpus-level IDF.

```python
CORPUS_VIDS = ["wbR-5mHI6bo", "BtN-goy9VOY", "0FH9cgRhQ-k", "JtUAAXe_0VI"]
VIDEO_ID = "udFxKZRyQt4"
LIMIT = 10000
TOP_X = 20
NGRAM_MIN = 1
NGRAM_MAX = 2
```

Scrape and preprocess videos first. Use subprocess allowing for-loop for CLI commands

```python
import subprocess

for vid in CORPUS_VIDS:
    print(f"Processing {vid}...")
    
    subprocess.run(
        [
            "py", "-m", "yt_comments",
            "scrape", vid,
            "--limit", str(LIMIT),
        ],
        check=True,
        capture_output=True,
        text=True
    )
    
    subprocess.run(
        [
            "py", "-m", "yt_comments",
            "preprocess", vid
        ],
        check=True,
        capture_output=True,
        text=True
    )

    print("Completed")
```

Build global corpus statistics from all available Silver artifacts.

```python
!py -m yt_comments corpus --ngram-min {NGRAM_MIN} --ngram-max {NGRAM_MAX} 
```

Scrape and preprocess video to be compared with global corpus.

```python
%%capture
!py -m yt_comments scrape {VIDEO_ID} --limit {LIMIT}
!py -m yt_comments preprocess {VIDEO_ID}
```

Run TF-IDF using global corpus.

```python
!py -m yt_comments tfidf {VIDEO_ID} --top-k {TOP_X} --ngram-min {NGRAM_MIN} --ngram-max {NGRAM_MAX} --use-corpus
```

Artifacts are now available under Bronze, Silver, and Gold in the `data/` folder.
